```python
#1.Remove only punctuation that separates sentences. 
text_string = origin_text.replace("/", ' ').replace(',', ' ').replace('?', ' ').replace('!', ' ').replace(':', ' ').strip()
```

```python
text_string
```

```python
#2. Extract name and email addresses  
emails = re.findall(r"[a-z0-9\.\-+_]+@[a-z0-9\.\-+_]+\.[a-z]+", origin_text)
print(emails)
```

```python
#3.Split by delimiter \n ,\t and whitespace then lowercase all tokens. 
processed_token = [i for i  in re.split('\s|\n|\t', text_string.strip()) if i != '']
processed_token
```

```python
#4. Remove tokens that don't contain any alphanumeric characters: 
#meaningless token such as [...] will be removed. 
processed_token = list(filter (lambda s:any([c.isalnum() for c in s]), processed_token))
processed_token
```

```python
#5. Clean out leading and trailing punctuation. 
processed_token = [i.rstrip('.').
                         rstrip(',').
                         lstrip('(').  
                         rstrip(')').
                         lstrip('<').
                         rstrip('>').
                         rstrip(':').
                         lstrip('>').
                         rstrip('>').
                         lstrip('"').
                         rstrip('"').
                         lstrip("'").
                         rstrip("'").
                         lstrip("-").
                         rstrip("-") for i in processed_token]
processed_token
```

```python
#6. Remove stopwords 
import nltk
from nltk.corpus import stopwords
import string
nltk.download('stopwords')
base_stop_words = stopwords.words("english")
punctuation_string = list(string.punctuation)
base_stop_words.extend(punctuation_string)   # add punctuation list to stopwords dictionary

extra_stopwords_black = {'n/a', 'unspecified',"subject"}  # add extra stopwords
extra_stopwords_white = {'build'}           # disable certain stopwords

stopwords_eng = set(base_stop_words)
stopwords_eng.update(extra_stopwords_black) 
stopwords_eng.difference_update(extra_stopwords_white)
processed_token = ' '.join([i for i in processed_token if i not in stopwords_eng])
processed_token
```
